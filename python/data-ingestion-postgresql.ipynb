```python
#Libs
import pandas as pd
from faker import Faker
from random import uniform, choice, randrange
from sqlalchemy import create_engine
from sqlalchemy.sql import text as sa_text

print('success')
```

```python
#create calendar table
calendar = []
calendar = [d.strftime('%Y-%m-%d') for d in pd.date_range(start="2019-01-01",end="2023-05-07")]

print('success')
```

```python
#list calendar table
df_calendar = pd.DataFrame(calendar,columns=['date'])
df_calendar
```

```python
#Create fake data
fake = Faker('pt-br')

#Create struct manager x coordinator
manager = []
coordinator = []

#create 10 manager
for idmanager in range(10):
    manager.append([idmanager, fake.name()])
    
    #create 10 coordinator by manager
    for idcoordinator in range(10):
        coordinator.append([idcoordinator + (idmanager*10), fake.name(),idmanager])

print('success')
```

```python
#list manager
df_manager = pd.DataFrame(manager,columns=['id','name'])
df_manager
```

```python
#list coordinator
df_coordinator = pd.DataFrame(coordinator,columns=['id','name','id_manager'])
df_coordinator
```

```python
#create strucs product x category
category = [[0,'Mobile phones'],[1,'Game consoles'],[2,'Household furniture'],[3,'Home appliances'],[4, 'Clothing']]
product = []

#create 5 products by category
for idcategory in category:
    for idproduct in range(10):
        product.append([idproduct + (idcategory[0]*10), fake.bothify(text='????-########').upper(), uniform(2.5, 10.0), idcategory[0]])

print('success')
```

```python
#list category
df_category = pd.DataFrame(category,columns=['id','name'])
df_category
```

```python
#list product
df_product = pd.DataFrame(product,columns=['id','name','price','id_category'])
df_product
```

```python
#Create struct customer
customer = []

#create 100 customers
for idcustomer in range(100):
    customer.append([idcustomer, fake.name()])
    
print('success')
```

```python
#list customers
df_customers = pd.DataFrame(customer,columns=['id','name'])
df_customers
```

```python
#Create struct orders with 5 milions rows
orders = []
for idorder in range(5000000):
    selected_product = choice(product)
    qty = randrange(1, 50)
    orders.append([idorder, choice(calendar), choice(customer)[0],selected_product[0],choice(coordinator)[0], qty, selected_product[2], qty * selected_product[2]])
    
print('success')
```

```python
#list orders
df_orders = pd.DataFrame(orders,columns=['id','date','id_customer','id_product','id_coordinator','qty','price_unit','total'])
df_orders
```

```python
#Connection
engine = create_engine('postgresql://postgres:1234@localhost:5432/SalesDW')

# Insert dataframe to local postgres (replace truncate and insert)
df_orders.to_sql('orders', engine, if_exists='replace')
df_customers.to_sql('customer', engine, if_exists='replace')
df_product.to_sql('product', engine, if_exists='replace')
df_category.to_sql('category', engine, if_exists='replace')
df_coordinator.to_sql('coordinator', engine, if_exists='replace')
df_manager.to_sql('manager', engine, if_exists='replace')
df_calendar.to_sql('calendar', engine, if_exists='replace')

print('success')
```
